```python
d = {"ineuron":{"a":1,"b":2,"c":3},"couse":{"d":45,"e":34,"f":8},"g":34,"h":[45,8,7,9,1,10],"i":(45,32,33),"k":"nikshit"}
```

```python
l = []
for i in d.values():
    print(i)
    
    if type(i)==dict:
        for j in i.values():
            print(j)
            l.append(j)
    elif type(i)==int:
        l.append(i)
        
    elif type(i)==list:
        l.extend(i)
        
    elif type(i)==tuple:
        for j in i:
            l.append(j)
print(l)

max=0
for i in l:
    if i>=max:
        max = i
        
print(max)
```
